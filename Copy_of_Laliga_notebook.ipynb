```
data_new["GoalsFor"] = pd.to_numeric(data_new["GoalsFor"])
data_new.dtypes
```

```
data_new["GoalsAgainst"] = pd.to_numeric(data_new["GoalsAgainst"])
data_new.dtypes
```

```
def Goal_diff_count(GoalsFor,GoalsAgainst):
  Goal_diff_count = GoalsFor - GoalsAgainst
  return(Goal_diff_count)
data_new['NewCol'] = data_new.apply(lambda x: Goal_diff_count(x['GoalsFor'], x['GoalsAgainst']), axis=1)
```

```
data_new["GamesWon"] = pd.to_numeric(data_new["GamesWon"])
```

```
data_new["GamesPlayed"] = pd.to_numeric(data_new["GamesPlayed"])
data_new.dtypes
```

```
def win_percent_count(GamesWon,GamesPlayed):
  if GamesWon !=0 and GamesPlayed !=0:
    Per_Winning = (GamesWon / GamesPlayed)*100
  else:
    Per_Winning = 0
  return(Per_Winning)
data_new['Winning Percent'] = data_new.apply(lambda x: win_percent_count(x['GamesWon'], x['GamesPlayed']), axis=1)
```
