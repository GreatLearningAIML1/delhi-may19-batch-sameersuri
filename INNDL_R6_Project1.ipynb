```
X = data.iloc[:, 3:13].values # customer names and ID have no effect on exit. Putting in an array except target column
print(X)
```

```
y = data.iloc[:, 13].values # Exit column
print(y)
```

```
from sklearn.preprocessing import LabelEncoder, OneHotEncoder # Gender and Geography need to convert into numerical data
```

```
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
```

```
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
```

```
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```
print(X)
```

```
from sklearn.model_selection import train_test_split
```

```
trainX,testX,trainY,testY = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```
trainX = trainX.astype('float32')
trainY = trainY.astype('float32')
```

```
testX = testX.astype('float32')
testY = testY.astype('float32')
```

```
trainX.shape
```

```
testX.shape
```

```
testY[0]
```

FEATURE SCALING

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
trainX = sc.fit_transform(trainX)
testX = sc.transform(testX)
```

```
testY[0]
```

## Build the Graph

```
#Initialize Sequential model
model = tf.keras.models.Sequential()

#Normalize the data
#model.add(tf.keras.layers.BatchNormalization())
```

## Build the Graph...

```
#Add 1st hidden layer
model.add(tf.keras.layers.Dense(6, input_shape=(11,),activation='sigmoid'))
```

```
#Add 2nd hidden layer
model.add(tf.keras.layers.Dense(6, activation='sigmoid'))
```

```
#Add 3rd hidden layer
model.add(tf.keras.layers.Dense(3, activation='sigmoid'))
```

```
#Add 4th hidden layer
#model.add(tf.keras.layers.Dense(30, activation='sigmoid'))
```

```
#Add the Dense Layer for Prediction
model.add(tf.keras.layers.Dense(1, input_shape=(11,),activation="sigmoid"))
```

## Build the Graph...

```
#Compile the model
model.compile(optimizer='sgd', loss='binary_crossentropy', 
              metrics=['accuracy'])
```

## Review model

```
model.summary()
```

## Train the model

```
model.fit(trainX,trainY,          
          validation_data=(testX,testY),
          epochs=100,
          batch_size=10)
```
